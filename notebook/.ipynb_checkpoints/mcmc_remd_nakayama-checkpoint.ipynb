```julia
using PyPlot, StatsBase, Printf, DelimitedFiles;
using Revise;
using MDToolbox;
PyPlot.plt.style.use("seaborn-colorblind");
ENV["COLUMNS"] = 110; #display width for MDToolbox
```

### Potential energy function

```julia
V(x; k=1.0) = sum(- (1.0 / 2.0) .* k .* x.^2 .+ (1.0 ./ 4.0) .* k .* x.^4)
V([0.0], k=1.0)
```

### Proposal distribution

```julia
proposal_fun(x) = x .+ (rand(Float64, size(x)) .* 2.0 .- 1.0);
proposal_fun([0.0])
```

### Replica MCMC without exchange

```julia
nreplica = 4
nstep = 100;
nexchange = 1000;
temperature_array = [0.01, 0.10, 0.30, 0.40];



x_array = []
for ireplica = 1:nreplica
    x = zeros(Float64, nstep);
    x[1:1] .= 0.0;
    push!(x_array, x)
end

for iexchange = 1:nexchange
    for ireplica = 1:nreplica
        x_array[ireplica][istep:istep] .= propagate_mcmc(y -> V(y, k=1.0), proposal_fun, x_array[ireplica][(istep-1):(istep-1)], temperature_array[ireplica], nstep=nstep);
    end
    # do not exchange
end
```

```julia
fig, ax = subplots(figsize=(8, 6))
for ireplica = 1:nreplica
    ax.plot(x_array[ireplica], linewidth=1.0)   
end
xlabel("step",fontsize=20)
ylabel("x(step)",fontsize=20)

ax.legend(["replica 1", "replica 2", "replica 3", "replica 4"])
tight_layout()
```

### Replica exchange MCMC

```julia
function exchange_temperature!(pot_fun::Function, icount, x_array, tempid2repid, repid2tempid, temperature_array)
    iaccepted = 0
    if mod(icount, 2) == 0
        for itemp = 1:2:(nreplica-1)
            repid_1 = tempid2repid[itemp]
            repid_2 = tempid2repid[itemp+1]
            x_1 = x_array[repid_1]
            x_2 = x_array[repid_2]
            delta=(beta_m[tmp_i[i_high]]-beta_m[tmp_i[i_low]])*(v_t[i_low]-v_t[i_high])
            delta = ((1.0 / temperature_array[itemp]) - (1.0 / temperature_array[itemp+1])) .* (pot_fun(x_array[x_2]) -  pot_fun(x_array[x_1]))
            if exp(- delta) > rand()
                tempid2repid[itemp], tempid2repid[itemp+1] = tempid2repid[itemp+1], tempid2repid[itemp]
                iaccepted += 1
            end
        end
    else
        for itemp = 2:2:(nreplica-1)
        end
    end
end
```

```julia
nreplica = 4
nstep = 100;
nexchange = 1000;
temperature_array = [0.01, 0.10, 0.30, 0.40];

x_array = []
for ireplica = 1:nreplica
    x = zeros(Float64, nstep);
    x[1:1] .= 0.0;
    push!(x_array, x)
end

tempid2repid = collect(1:nreplica);

for iexchange = nexchange
    for ireplica = 1:nreplica
        x_array[ireplica][istep:istep] .= propagate_mcmc(y -> V(y, k=1.0), proposal_fun, x_array[ireplica][(istep-1):(istep-1)], temperature_array[ireplica], nstep=nstep);
    end
end
```
